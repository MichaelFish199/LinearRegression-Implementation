This is one variable linear regresion using gradient descent\
In this code im not gonna use numpy(simplicity of lin alg) at all

```python
# Matplotlib for plotting
import matplotlib.pyplot as plt
```

```python
# Creating datapoints
x = [1,4,7,8,20]
y = [0,3,5,6,8]
n = len(x)
```

```python
# Algorythm variables
b_ = [] # lists to store how value changes
m_ = [] 

b = 0 # intercept
m = 0 # slope

a = 0.01 # learning rate
epochs = 2000 # num of iteration
```

```python
# Algorythm
for _ in range(epochs):
    sum1 = 0
    sum2 = 0

    for i in range(n):
        # Full batch gradient descent
        sum1 += ((b + m * x[i]) - y[i])
        sum2 += ((b + m * x[i]) - y[i]) * x[i]

    # Calculating derivarives
    t_b = b - a / n * sum1
    t_m = m - a / n * sum2
    
    # Updating values
    b = t_b
    m = t_m
    
    # saving values
    b_.append(b)
    m_.append(m)
```

```python
# Predicting
y_pred = [m * x[i] + b for i in range(len(x))]
```

```python
# Plotting regresion line
plt.scatter(x, y)
plt.plot(x, y_pred) # creating regresion line
plt.show()
```

```python
# Plotting how intercept changes
plt.plot(b_)
plt.show()
```

```python
# Plotting how slope changes
plt.plot(m_)
plt.show()
```
