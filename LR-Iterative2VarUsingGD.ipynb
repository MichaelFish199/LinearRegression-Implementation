This is two variable linear regresion using gradient descent\
In this code im not gonna use numpy(simplicity of lin alg) at all

```python
# Matplotlib for plotting
import matplotlib.pyplot as plt
```

```python
# Creating datapoints
x = [1,4,7,8,20]
z = [1,4,7,8,20]
y = [0,3,5,6,8]
```

```python
# Algorythm variables
n = len(x)

b = 0 # inicial intercept
m = 0 # inicial slope
g = 0 # inicial slope

a = 0.01 # learning rate
epochs = 1000 # num of iteration
```

```python
# Algorythm
for _ in range(epochs):
    sum1 = 0
    sum2 = 0
    sum3 = 0

    for i in range(n):
        sum1 += ((b + m * x[i] + g * z[i]) - y[i])
        sum2 += ((b + m * x[i] + g * z[i]) - y[i]) * x[i]
        sum3 += ((b + m * x[i] + g * z[i]) - y[i]) * z[i]
        
    # Calculating derivarives
    t_b = b - a / n * sum1
    t_m = m - a / n * sum2
    t_g = g - a / n * sum3
    
    # Updating values
    b = t_b
    m = t_m
    g = t_g
```

```python
# Predicting 
y_pred = [m * x[i] + g * z[i] + b for i in range(len(x))]
```
